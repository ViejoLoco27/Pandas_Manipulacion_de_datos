Se trata de un algoritmo capaz de analizar las características de una propiedad -como comodidades, tamaño, ocupación del espacio en un período determinado- y sugerir al anfitrión un precio a cobrar por tarifas diarias que garantice ganancias en momentos de alta demanda.

## **1. Importando librerías y datos**

```python
import pandas as pd
import numpy as np
```

```python
# Para leer el archivo se copio la ruta de acceso del documento
datos_hosting = pd.read_json('/content/datos_hosting.json')
datos_hosting
```

## **1.1. Normalizando datos**

```python
datos_hosting_normalize = pd.json_normalize(datos_hosting['info_inmuebles'])
datos_hosting_normalize
```

- 📋Las columnas presentadas en la tabla aún conservan algunos datos anidados en listas y en diccionarios.
- 📋Se aplicara el método explode para separar las columnas anidadas
- 📋Se aplicara el método reset.index para reiniciar el índice del nuevo DataFrame porque al separar las columnas anidadas por listas, los valores índices se replican.

```python
columnas = list(datos_hosting_normalize.columns)
columnas
```

```python
datos_hosting_normalize = datos_hosting_normalize.explode(columnas[3:])
datos_hosting_normalize
```

```python
datos_hosting_normalize.reset_index(drop=True, inplace=True)
datos_hosting_normalize
```

```python
datos_hosting_normalize.info()
```

# **2. Transformación de strings a datos numéricos**

```python
datos_hosting_normalize['max_hospedes'] = datos_hosting_normalize['max_hospedes'].astype(np.int64)
datos_hosting_normalize.info()
```

```python
col_numericas = ['cantidad_baños', 'cantidad_cuartos', 'cantidad_camas']
```

```python
datos_hosting_normalize[col_numericas] = datos_hosting_normalize[col_numericas].astype(np.int64)
```

```python
datos_hosting_normalize.info()
```

```python
datos_hosting_normalize['evaluacion_general'] = datos_hosting_normalize['evaluacion_general'].astype(np.float64)
```

```python
datos_hosting_normalize.info()
```

```python
datos_hosting_normalize['precio'] = datos_hosting_normalize['precio'].apply(lambda x: x.replace('$', '').replace(',','').strip())
```

```python
datos_hosting_normalize['precio'] = datos_hosting_normalize['precio'].astype(np.float64)
datos_hosting_normalize.info()
```

```python
datos_hosting_normalize[['cuota_deposito', 'cuota_limpieza']] = datos_hosting_normalize[['cuota_deposito', 'cuota_limpieza']].applymap(lambda x: x.replace('$', '').replace(',','').strip())
```

```python
datos_hosting_normalize
```

```python
datos_hosting_normalize[['cuota_deposito', 'cuota_limpieza']] = datos_hosting_normalize[['cuota_deposito', 'cuota_limpieza']].astype(np.float64)
```

```python
datos_hosting_normalize.info()
```

# **2.1. Manipulación de datos de strings**

```python
datos_hosting_normalize.head()
```

```python
datos_hosting_normalize['descripcion_local'] = datos_hosting_normalize['descripcion_local'].str.lower()
```

```python
datos_hosting_normalize['descripcion_local']
```

```python
datos_hosting_normalize['descripcion_local'][3169]
```

```python
#regex=True Expresión regular (palabra llave) que realiza una búqueda personalizada y más específica en el texto
# '[^a-zA-Z0-9\-\']´Estos son los caracteres que no (^) se van a modificar en el texto, todo lo demás sí
datos_hosting_normalize['descripcion_local'] = datos_hosting_normalize['descripcion_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
```

```python
datos_hosting_normalize['descripcion_local'][3169]
```

```python
#(?<!\w)-(?!\w) Estos caracteres indican que si hay caracteres antes o despúes de un guión no los modifique
datos_hosting_normalize['descripcion_local'] = datos_hosting_normalize['descripcion_local'].str.replace('(?<!\w)-(?!\w)', ' ', regex=True)
```

```python
datos_hosting_normalize['descripcion_local']
```

```python
datos_hosting_normalize['descripcion_local'] = datos_hosting_normalize['descripcion_local'].str.split()
datos_hosting_normalize['descripcion_local'].head()
```

```python
# Mediante el método split separa el string en una lista por cada palabra
datos_hosting_normalize['descripcion_local'][1]
```

```python
datos_hosting_normalize['comodidades'] = datos_hosting_normalize['comodidades'].str.replace('\{|}|\"','', regex=True)
```

```python
datos_hosting_normalize['comodidades']=datos_hosting_normalize['comodidades'].apply(lambda x: x.replace(' ', ','))
```

```python
datos_hosting_normalize['comodidades'] = datos_hosting_normalize['comodidades'].str.split(',')
```

```python
datos_hosting_normalize.head()
```

# **3. Deasfío de tokenizar datos**

Aprendimos a manipular elementos textuales con el comando str. Luego, usamos expresiones regulares para limpiar elementos no deseados en el texto y, finalmente, transformamos el texto tratado en una lista, construyendo un token.

Durante las clases transformamos dos columnas descripcion_local y comodidades buscando la tokenización. Pero aún falta la columna descripcion_vecindad, que también merece la pena transformar.

# **Actividad Adicional**

**Descripción:**

- convertir la columna precio adatos numéricos
- agrupar los precios respecto a las fechas para determinar en que fecha hubo más demanda

```python
datos_inmuebles.sample(15)
```

```python
# utilizamos el método fillna para llenar los elementos vacíos por '0.0'
# definimos el parámetro de inplace para True para substituir en el DataFrame
datos_inmuebles['precio'].fillna('0.0', inplace = True)
```

```python
datos_inmuebles['precio'].sample(15)
```

```python
datos_inmuebles['precio'] = datos_inmuebles['precio'].apply(lambda x: x.replace('$', '').replace(',','').strip())
datos_inmuebles.sample(15)
```

```python
datos_inmuebles['precio'] = datos_inmuebles['precio'].astype(np.float64)
datos_inmuebles.info()
```

```python
subset2 = datos_inmuebles.groupby(datos_inmuebles['fecha'].dt.strftime('%Y-%m'))['precio'].sum()
subset2 = subset2.sort_values(ascending=False)
```

```python
df = pd.DataFrame(subset2)
df
```
